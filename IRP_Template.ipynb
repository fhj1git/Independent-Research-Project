This study explores the best areas in Fife to locate wind turbines using Geographic Information Systems (GIS) to carry out a Multi-Criteria Evaluation (MCE) so as to inform policy-makers and energy developers. ## Introduction

Scotland is currently at the forefront globally, in terms of renewable energy, with a vast and growing renewable energy sector (Scottish Government, 2023). With a wealth of natural resources, Scotland has huge potential for renewable energy, particularly wind energy. Scotland alone makes up 39% of the UK's total renweable wind energy production and in 2020, supported over 10,000 jobs and added £1.2billion to the economy (Scottish Government, 2023). This has been identified as having the scope to grow to over 28,000 jobs directly and indirectly in the wind energy industry by 2030. Alongside this huge potential for economic growth from wind energy, exists the possibility to work towards the Scottish Government's aim of becoming net zero by 2045, and powering homes and businesses with zero carbon electricity (Scottish Government, 2021)(Scottish Government, 2023).

Inline with Government plans, Fife Council (2024) has identified a need to transform the energy system in Fife to one that is a low-carbon. However, Fife is currently constrained from transitioning to low-carbon energy system due to its existing infrastructure, hence the need for new infrastructure that would expand the capacity for renewable energy in Fife (Fife Council, 2024). With Scotland being such an important leader in terms of wind energy, it makes sense for Fife to build on this in its own efforts to transition to a low-carbon energy system. However, with multiple factors determining where to best locate a wind turbines, careful consideration is needed. Therefore, analysis was carried out to determine the best location for  on-shore wind turbines in Fife.

## Methodology
### Creating a Multi-Criteria Evaluation
In order to identify suitable areas for the location of new on-shore wind turbines, a Multi-Criteria Evaluation was carried out on QGIS. Seven key criteria were identified to deem a location suitable for wind turibines:


*   C1: The wind turbines should be located as close as possible to existing roads.
*   C2: The wind turbines cannot be loacted in protected areas.
*   C3: The wind turbines cannot be located on land with a slope in excess of 5.71°
*   C4: The wind turbines should be located in an area with as low a slope as possible
*   C5: The wind turbines should be located in an area with as high wind speeds as possible
*   C6: The wind turbines should be located in an area with as high a wind power density as possible
*   C7: The wind turbines cannot be located within 500m from the shoreline

C1 was determined a necessary factor, as road access is key to the accessibility of a wind turbine site, for construction and maintenance. Additionally, with a lack of electric grid data available, roads offer a good indication of the location of transmission lines. Thus a weighting of 20% was considered suitable, in line with the proportions of weightings determined by Shorabeh et al(2022). This data was accessed from Open Street Map (2024).

Determining and hence excluding protected areas from Fife (C2), was considered a crucially important constraint, due to legislation (Rankl, 2024) prohibiting the placement of wind turbines in these areas. 30x30 Protected Area data (Nature Scot, 2024), which includes all terrestrial protected areas in Scotland, was used to create this constraint.

A need for reasonably flat terrain determined C3 and C4. Blue Marble Geographics (n.d.) recommend areas with a slope less than 5.71° to be suitable for the sturdy construction of wind turbines (C3). Furthermore, the flatter the land the more suitable it is for wind turbine construction, due to greater foundation stability and ease of construction. A weighting attributed to the C4 factor of 20% was chosen in line with proportions of weighting determined by Shorabeh et al (2022). Scotland DEM5 (GG3209 Week 3, 2024) data, was converted to slope, and was used for both these criteria.

Key to the location of wind turbines, is the characteristics of the wind. In order for a wind turbine to generate power, sufficient windspeed is needed (C5), with higher windspeed correlating to greater power production (Blue Marble Geographics, n.d.). As no mean wind speed in Fife was above the cutoff of 24m/s, a constraint was not needed for this. High wind power density (C6); the mean annual power available per square meter of swept area (Taranovich, 2021), is key to determining where to best locate a wind farm for optimal energy production. Due to the essential role both of these factors play in the ability for wind turbines to produce power, they were both given a weighting of 30%. While the weighting of wind speed was based off of proportions of weighting by Shorabeh et al (2022), wind power was deemed to hold similar significance and hence given the same weighting. Data for both these criteria were sourced from the Global Wind Atlas (2024). The data was selected for 100m, as this was closest to the average height for a wind turbine (Blue Marble Geographics, n.d.).

A final constraint C7 was created to alleviate concerns of aesthetics. With the coast being one of the Fife's key areas for tourists, this would ensure the coastline was not aesthetically damaged. Additionally, with climate change predicted to contribute to sea levels rise and greater flood risk (UK Climate Risk, 2021), it was deemed sensible for the longevity of wind turbines to ensure they are not built too close to the shore. The data used for this was the Fife community council boundaries, merged to create the outline of Fife, from Fife Council(2014).


These criteria were then combined, considering the weights of each factor to create a suitability map with the following equation:

MCE = (C1x0.2 + C4x0.2 + C5 x 0.3 + C6 x 0.3) x C2 x C3 x C7

The raster file of the suitability map was then exported as a TIF to be used for further analysis. <br> <br>



###Visualising and Analysing
Using Google Collab, necessary Python libraries were installed and imported to allow for the visualisation and analysis of the MCE (view all in Appendix).

####Plotting suitability map
The raster file was loaded into Google Collab (using Rasterio library) and read as an array so that it could then be plotted. It was ensured that no-data values were masked so as to not obscure results, saving this as 'raster_data', which was then plotted using the Matplotlib library.

####Plotting distribution of suitability
The Earthpy library was then used to create a histogram of the array ('raster_data') and plotted with Matplotlib.


####Calculations
Numpy was used to calculate the total number of pixels in the 'raster_data' array. The size of the area of Fife was set to 1325km^2 (Office of National Statistics, 2024). These values were then used in the calculation of the percent and area size of Fife that fit certain suitability levels (see Appendix). The np.count_nonzero function was used for counting the number of pixels of certain suitability levels (this function counts all pixels of a specified value by setting all pixels of that value to 1 and all others to 0, then counting the total number of 1s). These values were then divided by the total number of pixels in Fife to find proportion of Fife for the different suitability levels. Percentages were then found by multiplying these values by 100. To find the area in km2 that each of these suitability levels covered, the proportion was multiplied by the area of Fife in km2. These values were then displayed in a table using Pandas and plotted using Matplotlib.

```python
with rio.open('/content/drive/MyDrive/Colab Notebooks/data/suitability map.tif') as suitability:
    suitability_arr = suitability.read(1)
    no_data = suitability.nodata

    if no_data is not None:
        raster_data = np.ma.masked_equal(suitability_arr, no_data)

plt.figure(figsize=(10, 8))
plt.imshow(raster_data, cmap='RdYlBu')
plt.colorbar(label='Suitability (0-1)', shrink=0.5)
plt.title("Figure 1 -Suitability Map for Wind Turbines in Fife", fontsize=11, pad=20)

plt.show()
```

There was found to be no areas of at least 80% suitability for the location of wind turbines in Fife (Figure 2).

## Discussion




```python
display(Image('/content/drive/MyDrive/Colab Notebooks/data/Fife council turbine map.png', width=800))
print("Figure 3 - Map of current and approved wind turbines in Fife. Source: Fife Plan (2019)")
```

Comparing areas from the MCE highlighted in the suitability map (Figure 1) against a map of already existing wind turbines (Figure 3) is hugely important for the validation of the MCE (Ffrench, 2021). Overlap indicates that the MCE is appropriately and accurately identifying areas in which wind turbines could be built. Focusing particularly on areas coloured in red, that indicate areas which are already wind turbine landscapes, for this comparison, these areas were found to largely match areas identified as suitable on the suitability map. While a turbine landscape currently exists in the south-east, our suitability map highlights the potential for expansion here to the north, south and west. The more central area highlighted in red in Figure 3 does not fully align with the area indicated to be suitable on the MCE. However, when the location of the  triangles, representing individual turbines, are considered, these line up more closely with areas highlighted as suitable on the map. This is possibly due to a focus in our analysis on only trunk, primary and motorway roads. Therefore, areas may have been deemed less suitable due to not being near roads when this may not reflect reality; they may be located near roads but just smaller ones. Comparison also highlights the north of the western border of Fife to still be undeveloped in terms of wind turbines, even though the MCE highlights this as a suitable area. While it would have been beneficial for the visualisation of this comparison to overlay existing turbines on the suitability map, this data is not publicly available in an appropriate format.


Although the chosen factors and constraints for this study offer a sufficient starting point for the identification of locations suitable for wind turbines in Fife, other studies suggest a range of further criteria necessary to identify upmost suitability. With the high costs associated with surveying areas, optimising suitability identification through MCE analysis is of great benefit to investors, as well as a saver of time and resources (Josimovic et al, 2023). Shorabeh et al (2022) advise the use of a criteria that considers the distance from water bodies. Proximity to water bodies threatens wind turbines with the risk of flooding. Although this is considered in terms of the shoreline, taking into account lakes and rivers could also be beneficial. While Ffrench (2021) suggest a constraint of 200m from water, Shorabeh et al (2022) suggest that areas within 1000m should be considered to have limited potential. Identifying a value within this range to create a constraint would help refine the MCE for Fife. Additionally,  vegetation is a further element not studied in the Fife MCE. Vegetation, such as woodland can inhibit the flow of wind, reducing the suitability of areas for the installation of wind turbines (Shorabeh et al, 2022). While this is somewhat accounted for with the consideration of wind speed and power in the MCE, consulting vegetation cover would help further refine the MCE. Using the Normalised Difference Vegetation Index (NDVI) would allow for the identification of highly vegetated areas which could then be removed from consideration. Josimovic et al (2023) also highlight the importance of considering legislation in areas in which the construction of wind turbines is prohibited to ensure relevant areas are excluded. This could include airport zones, national heritage areas and archaeological sites (Ffrench, 2021)(Josimovic et al, 2023). Further additional factors and constraints exist within the literature that could be considered to refine the MCE for wind turbines in Fife, in order to maximise efficiency and costs in the land identification process.

The MCE identified key areas of suitability within Fife for locating on-shorewind turbines. While a small area of Fife showed a high suitability of 70% and above, no area showed  an 80% or above suitability. This is due to the highly complex nature of determining a suitable area to locate wind turbines, as a result of many contesting factors and constraints. The MCE created offers a sufficient starting point for the consideration of a suitable locations for wind turbines in Fife, however, to progress further in identifying suitable sites it will be important to consider some additional key factors.
